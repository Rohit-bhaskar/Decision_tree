```python
import pandas as pd
df=pd.read_csv("Tennis.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_outlook=LabelEncoder()
le_temp=LabelEncoder()
le_humid=LabelEncoder()
le_wind=LabelEncoder()
```

```python
df["Outlook_"]=le_outlook.fit_transform(df["Outlook"])
df["Temperature_"]=le_temp.fit_transform(df["Temperature"])
df["Humidity_"]=le_humid.fit_transform(df["Humidity"])
df["Wind_"]=le_wind.fit_transform(df["Wind"])
```

```python
df
```

```python
df.drop(["Outlook","Temperature","Humidity","Wind"],axis=1,inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
target=df["Playtennis"]
```

```python
target
```

```python
df.drop("Playtennis",axis=1,inplace=True)
```

```python
df
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(df,target)
```

```python
model.score(df,target)
```
